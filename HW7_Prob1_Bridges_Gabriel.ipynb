```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1,3,100)
```

```python
Y = 3*(X**2)
```

```python
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Y=$3x^2$")
plt.show()
```
