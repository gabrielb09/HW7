```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t ,s, ds = np.loadtxt("Data7.csv",skiprows=4 , unpack=True, delimiter=",")
```

```python
velT = np.linspace(0.25,9.75,20)
v = (np.delete(s,0)-np.delete(s,20))/0.5
dv = 4*np.delete(ds,0)
```

```python
accT = np.linspace(0.5,9.5,19)
a = (np.delete(v,0)-np.delete(v,19))/0.5
da = 4*np.delete(dv,0)
```

```python
plt.figure(1,figsize=(10, 10))

plt.subplot(3,1,1)
plt.errorbar(t,s,yerr=ds,capsize=3.5,elinewidth=1)
plt.xlim(0,10)
plt.xlabel('time (s)')
plt.ylabel('position (m)')

plt.subplot(3,1,2)
plt.errorbar(velT,v,yerr=dv,capsize=3.5,elinewidth=1)
plt.xlim(0,10)
plt.xlabel('time (s)')
plt.ylabel('velocity ($\\frac{m}{s}$)')

plt.subplot(3,1,3)
plt.errorbar(accT,a,yerr=da,capsize=3.5,elinewidth=1)
plt.xlim(0,10)
plt.xlabel('time (s)')
plt.ylabel('acceleration ($\\frac{m}{s^2}$)')

plt.show()
```
