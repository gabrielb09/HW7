```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi,np.pi,100)
```

```python
Ycos = np.cos(X)
Ysin = np.sin(X)
```

```python
plt.plot(X,Ycos,'k',label='cos')
plt.plot(X,Ysin,'g',label='sin')
plt.legend()
plt.show()
```
