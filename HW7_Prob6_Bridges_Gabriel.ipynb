```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d=np.array([0.38, 0.64, 0.91, 1.26, 1.41, 1.66, 1.90, 2.18])
f=np.array([1.4, 1.65, 3.0, 3.95, 4.3, 5.20, 6.85, 7.4])
df=np.array([ 0.4, 0.5, 0.4, 0.5, 0.6, 0.5, 0.5, 0.4])
```

```python
yfit = np.array([0.5,8])
xfit = np.array([0.2,2.4])
yfitmax = np.array([0,8.5])
yfitmin = np.array([1,7.5])
```

```python
plt.errorbar(d,f,yerr=df,fmt = 'yd', ecolor = 'red',capsize=3.5,elinewidth=1,markeredgecolor='red',markersize=9)
plt.plot(xfit,yfit, 'k-', linewidth=1, label='m = 3.4$\pm$0.5')
plt.plot(xfit,yfitmax, 'k--', linewidth=1, alpha = 0.5)
plt.plot(xfit,yfitmin, 'k--', linewidth=1, alpha = 0.5)
plt.legend()
plt.xlim(0,2.5)
plt.ylim(0,9)
plt.xlabel('distance(cm)')
plt.ylabel('force(N)')
plt.show()
```
